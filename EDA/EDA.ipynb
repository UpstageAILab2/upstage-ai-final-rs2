```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. 데이터 읽기
file_path = '/data/ephemeral/home/upstage-ai-final-rs2/data/train.parquet'
df = pd.read_parquet(file_path)
```

```python
# 2. 기본 정보 확인
print("데이터프레임 기본 정보:")
print(df.info())

print("\n데이터프레임 요약 통계량:")
print(df.describe())

print("\n데이터프레임 상위 5개 행:")
print(df.head())
```

```python
# 3. 결측치 확인
print("\n결측치 확인:")
print(df.isnull().sum())
```

```python
# 4. 데이터 분포 확인
for column in df.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f'{column} Distribution')
    plt.show()
```
